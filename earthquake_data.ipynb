```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set a nice style for the plots
sns.set_style("whitegrid")

file_name = "earthquake_data_tsunami.csv"

try:
    # Load the dataset
    df = pd.read_csv(file_name)
    print(f"Successfully loaded '{file_name}' for visualization.")

    # --- Plot 1: Distribution of Earthquake Magnitudes ---
    plt.figure(figsize=(10, 5))
    sns.histplot(df['magnitude'], kde=True, bins=30)
    plt.title('Distribution of Earthquake Magnitudes')
    plt.xlabel('Magnitude')
    plt.ylabel('Frequency (Count)')
    print("\nGenerating Plot 1: Magnitude Distribution...")
    plt.show() # This displays the plot in Colab

    # --- Plot 2: Distribution of Earthquake Depths ---
    plt.figure(figsize=(10, 5))
    sns.histplot(df['depth'], kde=True, bins=30, color='red')
    plt.title('Distribution of Earthquake Depths (km)')
    plt.xlabel('Depth (km)')
    plt.ylabel('Frequency (Count)')
    print("Generating Plot 2: Depth Distribution...")
    plt.show()

    # --- Plot 3: Magnitude vs. Tsunami Event ---
    # This is a very important plot.
    # It compares the magnitudes of quakes that caused a tsunami (1)
    # vs. those that did not (0).
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=df, x='tsunami', y='magnitude')
    plt.title('Magnitude Comparison for Tsunami vs. Non-Tsunami Events')
    plt.xlabel('Tsunami Event (0 = No, 1 = Yes)')
    plt.ylabel('Magnitude')
    print("Generating Plot 3: Tsunami vs. Magnitude Boxplot...")
    plt.show()


    # --- Plot 7: Geographical Scatter Plot (Latitude vs. Longitude) ---
    # This plot shows the physical locations of the earthquakes.
    plt.figure(figsize=(12, 7))
    sns.scatterplot(data=df, x='longitude', y='latitude', hue='tsunami',
                    alpha=0.6, palette='Dark2')
    plt.title('Geographical Location of Earthquakes')
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.legend(title='Tsunami (1=Yes, 0=No)')
    print("\nGenerating Plot 7: Geographical Scatter Plot...")
    plt.show() # This displays the plot in Colab

    # --- Plot 8: Earthquakes by Month (Seasonality) ---
    # This bar chart shows if quakes are more common in certain months.
    plt.figure(figsize=(10, 5))
    sns.countplot(data=df, x='Month', palette='viridis')
    plt.title('Number of Earthquakes by Month')
    plt.xlabel('Month')
    plt.ylabel('Number of Earthquakes')
    print("Generating Plot 8: Earthquakes by Month...")
    plt.show()

    # --- Plot 9: Depth vs. Tsunami Event (Violin Plot) ---
    # A violin plot is like a boxplot, but also shows the
    # distribution (density) of the data on each side.
    plt.figure(figsize=(8, 6))
    sns.violinplot(data=df, x='tsunami', y='depth')
    plt.title('Depth Distribution for Tsunami vs. Non-Tsunami Events')
    plt.xlabel('Tsunami Event (0 = No, 1 = Yes)')
    plt.ylabel('Depth (km)')
    print("Generating Plot 9: Depth vs. Tsunami Violin Plot...")
    plt.show()

    print("\n--- Additional Visualizations Complete ---")

except FileNotFoundError:
    print(f"\nERROR: The file '{file_name}' was not found.")
    print("Please make sure you have uploaded the file using the Step 1 code.")
except Exception as e:
    print(f"An error occurred: {e}")
```
